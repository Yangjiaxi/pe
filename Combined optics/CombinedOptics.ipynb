```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
import seaborn as sb
```

```python
sb.set()
sb.set_style("whitegrid")
plt.rcParams['figure.figsize'] = [8, 4]
plt.rcParams['figure.dpi'] = 144
plt.rcParams['font.family']=['Arial Unicode MS']
```

```python
def il(str=None):
    if str is not None:
        res = []
        for ele in str.split('/'):
            res.append([float(e) for e in ele.split('+')])
        if len(res) == 1:
            return res[0]
        else:
            return res
    else:
        return None
```

```python
def make_ls(a, b, step):
    return np.linspace(a, b, int((b - a) / step + 1))
```

```python
x = make_ls(0, 20.4, 0.3)
```

```python
y = [il("0.122+0.117+0.106+0.111+0.123+0.146+0.197+0.302+0.394+0.407+0.483+0.525"),
il("0.527+0.492+0.452+0.359+0.244+0.164+0.102+0.076+0.230+0.509"),
il("0.661+1.272+2.17+2.72+3.56+4.78+5.93+7.18+8.38+9.17"),
il("10.06+10.30+10.45+10.41+10.00+9.25+8.38+7.61+6.31+4.91"),
il("4.14+3.05+2.01+1.36+0.869+0.445+0.214+0.093+0.124+0.226"),
il("0.287+0.364+0.424+0.455+0.463+0.438+0.384+0.313+0.236+0.192"),
il("0.145+0.090+0.085+0.090+0.120+0.134+0.166")]
```

```python
a = []
for line in y:
    a += line
a = np.array(a) - (-0.007)
```
